```python
TRAIN_CSV = "data/train/combined_train_dataset.csv"
VAL_CSV   = "data/val/combined_val_dataset.csv"
TEST_CSV  = "data/test/combined_test_dataset.csv"
MODEL_NAME = "facebook/bart-base"
MAX_LEN = 1024
SEED = 42
BATCH_SIZE = 8
GRAD_ACCUM = 2
EPOCHS = 2
LR = 3e-5

import os, math, numpy as np, pandas as pd, torch, random
from datasets import load_dataset, DatasetDict
from transformers import (AutoTokenizer, AutoModelForSequenceClassification,
                          DataCollatorWithPadding, TrainingArguments, Trainer)
from sklearn.metrics import mean_absolute_error, cohen_kappa_score, accuracy_score

random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
```

```python
from huggingface_hub import login
import wandb
!hf auth login --token "your_token"
!wandb login your_api_key

PROJECT = "bart-star-regression"
RUN_NAME = "bart-base_regression"

wandb.init(
    project=PROJECT,
    name=RUN_NAME,
    config={
        "model_name": "facebook/bart-base",
        "task": "regression",
        "max_len": MAX_LEN,
        "lr": LR,
        "epochs": EPOCHS,
        "batch_size": BATCH_SIZE,
        "grad_accum": GRAD_ACCUM,
        "seed": SEED,
    },
)
```

```python
raw = load_dataset(
    "csv",
    data_files={
        "train": TRAIN_CSV,
        "test": TEST_CSV,
        "val": VAL_CSV
    }
)

def keep_columns(ds):
    cols = ds.column_names
    needed = ["review_data", "rating"]
    to_remove = [c for c in cols if c not in needed]
    return ds.remove_columns(to_remove) if len(to_remove) else ds

raw = DatasetDict({k: keep_columns(v) for k, v in raw.items()})

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, use_fast=True)
tokenizer.padding_side = "right"

def preprocess(batch):
    # Tokenize text
    tokenized = tokenizer(
        batch["review_data"],
        max_length=MAX_LEN,
        truncation=True,
        padding=False  # we'll pad dynamically with the collator
    )
    # Labels as float for regression
    tokenized["labels"] = [float(x) for x in batch["rating"]]
    return tokenized

tokenized = DatasetDict({
    split: ds.map(preprocess, batched=True, remove_columns=ds.column_names)
    for split, ds in raw.items()
})
```

```python
model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=1,
    problem_type="regression"
)

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
output_dir = "bart_regression_star"

from transformers import TrainingArguments
args = TrainingArguments(
    output_dir=output_dir,
    overwrite_output_dir=True,
    logging_steps=100,
    save_steps=1000,
    per_device_train_batch_size=1,
    gradient_accumulation_steps=GRAD_ACCUM,
    num_train_epochs=EPOCHS,
    learning_rate=LR,
    seed=SEED,
    report_to="wandb",
)


trainer = Trainer(
    model=model,
    args=args,
    train_dataset=tokenized["train"],
    eval_dataset=tokenized["val"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=regression_metrics,
)
```

```python
train_result = trainer.train()
trainer.save_model(output_dir)
tokenizer.save_pretrained(output_dir)
print("Best checkpoint saved to:", output_dir)
```

```python
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer

# Reload model & tokenizer
model = AutoModelForSequenceClassification.from_pretrained(output_dir)
tokenizer = tokenizer.from_pretrained(output_dir)  # or AutoTokenizer

# Low-memory eval settings
eval_args = TrainingArguments(
    output_dir=output_dir,
    per_device_eval_batch_size=2,  # smaller batch size
    eval_accumulation_steps=64,    # accumulate preds on CPU
    dataloader_drop_last=False
)

# Recreate Trainer (no need for train_dataset)
trainer = Trainer(
    model=model,
    args=eval_args,
    tokenizer=tokenizer,
    compute_metrics=regression_metrics
)
```

```python
val_sample = tokenized["val"].shuffle(seed=42).select(range(min(1000, tokenized["val"].num_rows)))
val_full_metrics = trainer.evaluate(eval_dataset=val_sample)
print("\n=== Validation (full) ===")
for k, v in val_full_metrics.items():
    print(f"{k}: {v:.6f}" if isinstance(v, (int, float)) else f"{k}: {v}")

# Save full-val predictions
val_full_pred = trainer.predict(tokenized["val"])

# ✅ tuple-safe logits handling
logits = val_full_pred.predictions
if isinstance(logits, tuple):
    logits = logits[0]

y_true_full = val_full_pred.label_ids.astype(float)
y_pred_full = np.asarray(logits, dtype=float).squeeze()

df_full = pd.DataFrame({
    "true_rating": y_true_full,
    "pred_rating": y_pred_full,
    "pred_rating_rounded": np.rint(np.clip(y_pred_full, 1.0, 5.0)).astype(int)
})
full_path = os.path.join(output_dir, "val_full_predictions.csv")
df_full.to_csv(full_path, index=False)
print("Saved:", full_path)
```
